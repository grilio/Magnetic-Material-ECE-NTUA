```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt 
```

First we read the file as a pandas dataframe.

```python
values_file = "ΑΣΚΗΣΗ 1.xlsx"

df = pd.read_excel(values_file)
```

Now we select the columns where the data is and convert them into numpy arrays for practicallity. <br><br>
"to_numpy()" is used to convert dataframe to numpy array.<br>
"dropna()" is used to remove NaN values.

```python
H1 = df["H (A/m)"].dropna().to_numpy()
B1 = df["B (T)"].dropna().to_numpy()

H2 = df["H (A/m).1"].dropna().to_numpy()
B2 = df["B (T).1"].dropna().to_numpy()
```

```python
print("For the first material we have {} samples and for the second we have {} samples.".format(len(H1), len(H2)))
```

Now we can start the calculations.

#### Μ(Η): μαγνήτιση ως προς το εφαρμοζόμενο πεδίο

```python
m0 = 4 * np.pi * (10**(-7))
def M_calc(B, H):
    return (B / m0 - H)
M1 = M_calc(B1, H1)
M2 = M_calc(B2, H2)
```
